```python
import pandas as pd
import numpy as np
```

```python
def calculate_priors(data, target_column):
    classes = data[target_column].unique()
    priors = {}
    total = len(data)

    for c in classes:
        priors[c] = len(data[data[target_column]==c]) / total

    return priors
```

```python
def calculate_likelihood(data, feature, value, target_column, target_class):
    subset = data[data[target_column] == target_class]
    likelihood = len(subset[subset[feature] ==  value]) / len(subset)
    return likelihood
```

```python
def Naive_bayees(data, target_column, input_data):
    priors = calculate_priors(data, target_column)
    classes = priors.keys();

    posteriors = {}

    for c in classes:
        prob = priors[c]
        for feature, value in input_data.items():
            prob *= calculate_likelihood(data, feature, value, target_column, c);
        posteriors[c] = prob

    return max(posteriors, key = posteriors.get), posteriors
```

```python
data = pd.DataFrame({
    'Outlook' : ["Sunny", "Sunny", "Overcast", "Rain", "Rain", "Rain", "Overcast", "Sunny", "Sunny", "Rain"],
    'Humidity' : ["High", "High", "High", "High", "Normal", "Normal", "Normal", "High", "Normal", "Normal"],
    'Play' :["No", "No", "Yes", "Yes", "Yes", "No", "Yes", "No", "Yes", "Yes"]
})
```

```python
test_data = {'Outlook': 'Rain', 'Humidity': 'High'}
prediction, post = Naive_bayees(data, 'Play', test_data)
sum = post['No'] + post['Yes']
print("Probabilities: No - ",post['No']/sum,", Yes - ",post['Yes']/sum)
print("Prediction:", prediction)
```
